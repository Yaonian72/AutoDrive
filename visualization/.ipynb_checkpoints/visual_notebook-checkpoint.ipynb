```python
# set dataset
mmdet3d_dir = '../mmdetection3d/'
dataset = kitti_object(os.path.join(mmdet3d_dir, 'data/kitti'), 'training')
config_file = os.path.join(mmdet3d_dir, "configs/3dssd/3dssd_kitti-3d-car.py")
ckpt_file = os.path.join(mmdet3d_dir, "checkpoints/3dssd_kitti-3d-car/latest.pth")
```

```python
# get data
data_idx = 20
objects = dataset.get_label_objects(data_idx)
pc_velo = dataset.get_lidar(data_idx)
calib = dataset.get_calibration(data_idx)
img = dataset.get_image(data_idx)
img_height, img_width, _ = img.shape
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, ckpt_file, device='cuda:0')
result, data = inference_detector(model, dataset.get_lidar_filename(data_idx))
```

```python
# select bbox
threshold = 0.5
bboxes, scores = result[0]['boxes_3d'].corners, result[0]['scores_3d']
bboxes3d_pred = bboxes[scores > threshold]
bboxes3d_pred.shape
```

```python
# visualization
fig_3d = mlab.figure(bgcolor=(0, 0, 0), size=(800, 450))
mlab.clf()
draw_gt_boxes3d(bboxes3d_pred, fig_3d, color=(1, 0, 0))
show_lidar_with_depth(pc_velo, objects, calib, fig_3d, False, img_width, img_height)
fig_3d
```

```python
img_bbox2d, img_bbox3d = show_image_with_boxes(img, objects, calib)
img_bbox2d = cv2.cvtColor(img_bbox2d, cv2.COLOR_BGR2RGB)

for bbox3d in bboxes3d_pred:
    bbox2d = calib.project_velo_to_image(bbox3d)
    bbox2d = calib.project_8p_to_4p(bbox2d)
    cv2.rectangle(
            img_bbox2d,
            (int(bbox2d[0]), int(bbox2d[1])),
            (int(bbox2d[2]), int(bbox2d[3])),
            (255, 0, 0),
            2,
        )

fig_bbox2d = plt.figure(figsize=(14, 7))
ax_bbox2d = fig_bbox2d.subplots()
ax_bbox2d.imshow(img_bbox2d)
plt.imshow(img_bbox2d)
```
